## Reading All Datasets

```python
customers = pd.read_csv('/content/drive/MyDrive/Data Science/Assignment 3/olist_customers_dataset.csv')
geoLocation = pd.read_csv('/content/drive/MyDrive/Data Science/Assignment 3/olist_geolocation_dataset.csv')
orderItems = pd.read_csv('/content/drive/MyDrive/Data Science/Assignment 3/olist_order_items_dataset.csv')
orderPayments = pd.read_csv('/content/drive/MyDrive/Data Science/Assignment 3/olist_order_payments_dataset.csv')
orderReviews = pd.read_csv('/content/drive/MyDrive/Data Science/Assignment 3/olist_order_reviews_dataset.csv')
orders = pd.read_csv('/content/drive/MyDrive/Data Science/Assignment 3/olist_orders_dataset.csv')
products = pd.read_csv('/content/drive/MyDrive/Data Science/Assignment 3/olist_products_dataset.csv')
sellers = pd.read_csv('/content/drive/MyDrive/Data Science/Assignment 3/olist_sellers_dataset.csv')
categoryName = pd.read_csv('/content/drive/MyDrive/Data Science/Assignment 3/product_category_name_translation.csv')

# Expand the Output Display
pd.set_option('display.max_columns', 500)
```

```python
# Change Column Name in customers and geoLocation
customers.rename(columns={'customer_zip_code_prefix': 'zip_code'}, inplace=True)
geoLocation.rename(columns={'geolocation_zip_code_prefix': 'zip_code'}, inplace=True)
```

## Merging All Datasets Into One

```python
data = orders.merge(customers, 
                    on='customer_id').merge(orderItems, 
                                            on='order_id').merge(products, 
                                                                 on='product_id').merge(categoryName, 
                                                                                        on='product_category_name').merge(sellers, 
                                                                                                                          on='seller_id')

# Read Merged Datasets                                                                                                                                              
data
```

## Checking and Changing Data Types of Each Column

```python
data.dtypes
```

```python
# Changing Inappropriate Data Types 
data[  
     ['product_category_name', 
     'seller_city', 
     'customer_state', 
     'seller_state', 
     'customer_city']
     ] = data[
          ['product_category_name', 
          'seller_city', 
          'customer_state', 
          'seller_state', 
          'customer_city']
          ].astype('category')

data[
    ['order_purchase_timestamp']
     ]=data[
         ['order_purchase_timestamp']
            ].astype('datetime64')
```

```python
# Rechecking Data Types
data.dtypes
```

## Save and Export Datasets
